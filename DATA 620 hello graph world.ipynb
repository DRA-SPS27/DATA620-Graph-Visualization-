```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Set up an empty undirected graph
G = nx.Graph()

# 2. Define the nodes (the individuals in the kite)
nodes = [
    "Andre", "Beverly", "Carol", "Diane", "Ed", 
    "Fernando", "Garth", "Heather", "Ike", "Jane"
]
G.add_nodes_from(nodes)

# 3. Define the edges (the connections based on the Krackhardt kite diagram)
edges = [
    ("Andre", "Beverly"), ("Andre", "Carol"), ("Andre", "Diane"),
    ("Beverly", "Diane"), ("Beverly", "Ed"),
    ("Carol", "Diane"), ("Carol", "Fernando"),
    ("Diane", "Ed"), ("Diane", "Fernando"), ("Diane", "Garth"),
    ("Ed", "Garth"),
    ("Fernando", "Garth"), ("Fernando", "Heather"),
    ("Garth", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
]
G.add_edges_from(edges)

# 4. Visualize the graph
plt.figure(figsize=(10, 6))

# Use a spring layout for aesthetically pleasing node spacing
pos = nx.spring_layout(G, seed=42) 

# Draw the graph components
nx.draw(
    G, pos, 
    with_labels=True, 
    node_color='#DDA0DD', # Plum/purple color 
    node_size=2000, 
    font_size=12, 
    font_weight='bold', 
    edge_color='purple', 
    width=2
)

plt.title("Hello, Graph World: The Krackhardt Kite")
plt.show()
```
